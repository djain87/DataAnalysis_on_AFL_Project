```python
# Perform an inner join based on a common column
merged_df = games_df.merge(stats_df, on='gameId')

# Print the merged dataframe
merged_df.info()
```

```python
#Find average goals scored in Home matches for each team

home_average_score = merged_df.groupby("team")["homeTeamScore"].mean()
print(home_average_score)
```

```python
#Find average goals scored in Away matches for each team
away_average_score = merged_df.groupby("team")["awayTeamScore"].mean()
print(away_average_score)
```

```python
#Find the win rate for Home vs Away for each team

#Create a new column 'Result' to indicate the result of each match based on home and away scores.
merged_df["result"] = np.where(merged_df["homeTeamScore"]>merged_df["awayTeamScore"],"homeWin","awayWin")

#Group the DataFrame by 'homeTeam' and 'Result' columns and count the occurrences.
win_count = merged_df.groupby(['homeTeam','result']).size().reset_index(name='Count')

#Pivot the data to have separate columns for home and away wins.
win_rate = win_count.pivot(index='homeTeam', columns='result', values='Count')

#Calculate the win rate for each team.
win_rate['WinRate_Home'] = win_rate['homeWin'] / (win_rate['homeWin'] + win_rate['awayWin'])
win_rate['WinRate_Away'] = win_rate['awayWin'] / (win_rate['homeWin'] + win_rate['awayWin'])

#Print the resulting win rate for Home vs Away for each team.
print(win_rate)
```

```python
#Which team won the maximum games back to back

#Create a new column 'WinStreak' to track the consecutive win streak for each team.
merged_df['WinStreak'] = merged_df.groupby('homeTeam')['result'].apply(lambda x: x.eq('HomeWin').cumsum().where(x.eq('HomeWin'), 0))

#Group the DataFrame by 'homeTeam' and 'WinStreak' columns and count the maximum win streak for each team
max_streak = merged_df.groupby(['homeTeam', 'WinStreak']).size().reset_index(name='Count')

#Sort the DataFrame by the 'Count' column in descending order to find the maximum win streak for each team
max_streak = max_streak.sort_values('Count', ascending=False)

print(max_streak)
#Get the team with the maximum win streak
team_with_max_streak = max_streak.iloc[0]['homeTeam']

#Print the team that won the maximum number of games back to back
print("Team with maximum consecutive win streak:", team_with_max_streak)
```

```python
#What were the top 5 years with the most games held?
#Group the DataFrame by the 'year_x' column and count the number of games in each year.
games_per_year = merged_df.groupby("year_x").size().reset_index(name="gamecount")

#Sort the DataFrame in descending order based on the 'GameCount' column.
games_per_year.sort_values("gamecount", ascending = False)

#Retrieve the top 5 years with the most games.
top_5_years = games_per_year.head(5)['year_x']


#Print the top 5 years with the most games held.
print("Top 5 years with the most games held:")
for year in top_5_years:
    print(year)
    
#Based on the top years, which 3 venues had the most games held during this period?     
filtered_df = merged_df[merged_df["year_x"].isin(top_5_years)]

#Group the filtered DataFrame by the 'venue' column and count the number of games held at each venue
venue_games_count = filtered_df.groupby('venue').size().reset_index(name='GameCount')

#Sort the DataFrame in descending order based on the 'GameCount' column
venue_games_count = venue_games_count.sort_values('GameCount', ascending=False)

#Retrieve the top 3 venues with the most games held
top_3_venues = venue_games_count.head(3)['venue']

#Print the top 3 venues with the most games held.
print("\n""Top 3 venues with the most games held during the top years:")
for venue in top_3_venues:
    print(venue)
```

```python
#What were the bottom 5 years with the least games held?  

bottom_5_years = games_per_year.tail(5)['year_x']

#Print the bottom 5 years with the most games held.    
print("\n""Bottom 5 years with the most games held:")
for year in bottom_5_years:
    print(year)
```

```python
#Does the distribution for the crowd in attendance follow a normal distribution? If not briefly describe the type of distribution. 

import matplotlib.pyplot as plt
from scipy import stats

# Create a histogram of crowd attendance
plt.hist(merged_df["attendance"], bins=20, density=True, alpha=0.5, label='Crowd Attendance')

# Calculate mean and standard deviation
mean = np.mean(merged_df["attendance"])
std = np.std(merged_df["attendance"])

# Generate values for the normal distribution curve
x = np.linspace(min(merged_df["attendance"]), max(merged_df["attendance"]), 100)
y = stats.norm.pdf(x, mean, std)

# Overlay the normal distribution curve on the histogram
plt.plot(x, y, 'r-', label='Normal Distribution')

# Add labels and legend
plt.xlabel('Crowd Attendance')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```

```python
#The Shapiro-Wilk test is a statistical test used to assess the normality of a dataset. 
#It determines whether a given sample of data comes from a normally distributed population.
#The test is based on the comparison of observed data with the expected values under the assumption of normality.
#The Shapiro-Wilk test calculates a test statistic (W) based on the sample data and provides a p-value.
#The test evaluates the null hypothesis that the data follows a normal distribution.
#If the p-value is greater than a chosen significance level (commonly 0.05), the null hypothesis is not rejected,
#suggesting that the data can be reasonably assumed to be normally distributed.
#However, if the p-value is less than the significance level,
#the null hypothesis is rejected, indicating that the data significantly deviates from a normal distribution


from scipy import stats

# Perform Shapiro-Wilk test
statistic, p_value = stats.shapiro(merged_df["attendance"])

print(p_value)
# Compare p-value with significance level
if p_value > 0.05:
    print("The crowd attendance data follows a normal distribution.")
else:
    print("The crowd attendance data does not follow a normal distribution.")
```
